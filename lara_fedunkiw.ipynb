```python
# En el siguiente ejercicio se calcula el area de un rectangulo

def rectangle_area(width, height):
    area = width * height
    print(f"The area of a rectangle with width {width} and height {height} is {area}") 
```

```python
def central_difference(f, x, h):
    derivada_aproximada = (f(x + h) - f(x - h)) / (2 * h)

    return derivada_aproximada
```

```python
def square(x):
    return x**2
y = square
x = 2
h1 = 0.1
h2 = 0.01
h3 = 0.001

print(f"the derivative of f(x) = x^2 at x = {x} using h = {h1} is {central_difference(y, x, h1)}.")
print(f"the derivative of f(x) = x^2 at x = {x} using h = {h2} is {central_difference(y, x, h2)}.")
print(f"the derivative of f(x) = x^2 at x = {x} using h = {h3} is {central_difference(y, x, h3)}.")
```

##### the derivative of f(x) = x^2 at x = 2 using h = 0.1 is 4.100000000000001.<br>the derivative of f(x) = x^2 at x = 2 using h = 0.01 is 4.009999999999891.<br>the derivative of f(x) = x^2 at x = 2 using h = 0.001 is 4.000999999999699.
