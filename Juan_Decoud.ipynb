# [EXERCISE 5]

```python
def central_difference(f, x, h):
    return (f(x+h)-f(x))/h
```

```python
def square(x):
    return x**2
f = square
x = 2
h1 = 0.1
h2 = 0.01
h3 = 0.001

print(f"The derivative of f(x) = x^2 at x = {x} using h = {h1} is {central_difference(f, x, h1)}.")

print(f"The derivative of f(x) = x^2 at x = {x} using h = {h1} is {central_difference(f, x, h2)}.")

print(f"The derivative of f(x) = x^2 at x = {x} using h = {h1} is {central_difference(f, x, h3)}.")
```
