```python
#Write a function central_difference(f, x, h) that approximates the derivative of a function f at a point x using the Central Difference Approximation. The function should take the function f, the point at which the derivative needs to be approximated x, and the step size h as input, and return the approximate value of the derivative.

def central_difference(f, x, h):
    derivative = (f(x + h) - f(x - h)) / (2 * h)
    return derivative
```

```python
def f(x):
    return x ** 2
x = 2
h = 0.1

approx_derivative_1 = central_difference(f, x, h)
print("The derivative of f(x) = x^2 at x = 2 using h = 0.1 is:", approx_derivative_1)

h = 0.01

approx_derivative_2 = central_difference(f, x, h)
print("The derivative of f(x) = x^2 at x = 2 using h = 0.01 is:", approx_derivative_2)

h = 0.001
approx_derivative_3 = central_difference(f, x, h)
print("The derivative of f(x) = x^2 at x = 2 using h = 0.001 is:", approx_derivative_3)
```
