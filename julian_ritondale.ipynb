```python
def central_difference(f, x, h):
    a = f
    function = f
    v1 = x + (h/2)
    v2 = x - (h/2)
    f = list(f)
    function = list(function)
    for var in f:
        if var == 'x':
            f[f.index(var)] = f'{v1}'
    for var in function:
        if var == 'x':
            function[function.index(var)] = f'{v2}'
    f = ''.join(f)
    function = ''.join(function)
    approximation = (eval(f) - eval(function))/h
    return f"The derivative of f(x) = {a} at x = {x} using h = {h} is {approximation}."
```

```python
print(central_difference('x**2',2,0.1))
print(central_difference('x**(-3/2) + 5*(x**2) - 10*x + 2',10,0.01))
print(central_difference('x**-1',5,0.001))

```
