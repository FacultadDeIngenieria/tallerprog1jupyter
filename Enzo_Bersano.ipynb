```python

def central_difference(f, x, h):
     return ((f(x+h)-f(x))/h)
```

```python

def square(x):
    return x**2
y = square
x = 2
h1 = 0.1
h2 = 0.01
h3 = 0.001

print(f"the derivative of f(x) = x^2 at x = {x} using h = {h1} is {central_difference(y, x, h1)}.")
print(f"the derivative of f(x) = x^2 at x = {x} using h = {h2} is {central_difference(y, x, h2)}.")
print(f"the derivative of f(x) = x^2 at x = {x} using h = {h3} is {central_difference(y, x, h3)}.")
```
