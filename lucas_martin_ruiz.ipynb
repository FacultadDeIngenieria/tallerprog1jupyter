```python
def central_difference(f, x, h):
    x1 = x+(h)

    return (f(x1)-f(x))/h
```

```python
def f(x):
    return x**2
print(f"The derivative of f(x) = x^2 at x = 2 using h = 0.1 is {central_difference(f, 2, 0.1)}")
print(f"The derivative of f(x) = x^2 at x = 2 using h = 0.01 is {central_difference(f, 2, 0.01)}")
print(f"The derivative of f(x) = x^2 at x = 2 using h = 0.001 is {central_difference(f, 2, 0.001)}")
```
