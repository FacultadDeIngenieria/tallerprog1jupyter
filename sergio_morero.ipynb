# [EXERCISE 2]

```python
#Esta función se encarga de dar el perímetro de un rectángulo tras haberle dado su ancho y su alto como datos.
def rectangle_area(width, height):
    return width * height
```

# [EXERCISE 3]

```python
#La funcion previamente definida recibe 2 valores (ancho y alto) y devuelve, en cada caso, su perimetro correspondiente.
print(f"The area of a rectangle with width 3 and height 4 is {rectangle_area(3, 4)}")
print(f"The area of a rectangle with width 5 and height 10 is {rectangle_area(5, 10)}")
print(f"The area of a rectangle with width 8 and height 2 is {rectangle_area(8, 2)}")
```

# [EXERCISE 5]

```python
def central_difference(f, x, h):
    return (f(x+h)-f(x))/h
```

```python
def square(x):
    return x**2
f = square
x = 2
h1 = 0.1
h2 = 0.01
h3 = 0.001
print(f"The derivative of f(x) = x^2 at x = {x} using h = {h1} is {central_difference(f, x, h1)}.")
print(f"The derivative of f(x) = x^2 at x = {x} using h = {h2} is {central_difference(f, x, h2)}.")
print(f"The derivative of f(x) = x^2 at x = {x} using h = {h3} is {central_difference(f, x, h3)}.")
```
